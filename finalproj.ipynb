```python
X_train=X_train.head(4000)
X_test=X_test.head(4000)
```

```python
x_train=X_train.drop(['train_target'],axis=1)
y_train=X_train['train_target']	
x_test=X_test.drop(['test_target'],axis=1)
y_test=X_test['test_target']
```

```python
#Scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression



model=LinearRegression()
model.fit(x_train_scaled,y_train)
train_pred=model.predict(x_train_scaled)
test_pred=model.predict(x_test_scaled)
print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")
print(f"Train MAE: {mean_absolute_error(y_train,train_pred)}")
print(f"Test MAE: {mean_absolute_error(y_test,test_pred)}")

r2=r2_score(y_test,test_pred)


print("R squared value:",r2)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.svm import SVR
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report




model1=SVR()

model1.fit(x_train,y_train)
train_pred=model1.predict(x_train)
test_pred=model1.predict(x_test)
print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")
print(f"Train MAE: {mean_absolute_error(y_train,train_pred)}")
print(f"Test MAE: {mean_absolute_error(y_test,test_pred)}")


r2=r2_score(y_test,test_pred)


print("R squared value:",r2)
#print(classification_report(y_test,test_pred))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score


model2=Ridge()
model2.fit(x_train_scaled,y_train)
train_pred=model2.predict(x_train_scaled)
test_pred=model2.predict(x_test_scaled)
print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")


r2=r2_score(y_test,test_pred)


print("R squared value:",r2)

```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score


model3=Lasso()
model3.fit(x_train_scaled,y_train)
train_pred=model3.predict(x_train_scaled)
test_pred=model3.predict(x_test_scaled)
print(f"Train MSE: {mean_squared_error(y_train,train_pred)}")
print(f"Test MSE: {mean_squared_error(y_test,test_pred)}")


r2=r2_score(y_test,test_pred)


print("R squared value:",r2)
```
